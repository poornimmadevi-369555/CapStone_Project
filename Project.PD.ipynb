```python
X = df.drop("Class", axis=1)
y = df["Class"]

print("Features:", X.shape)
print("Target:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
print(X_train.shape)
print(y_train.shape)
print(type(X_train))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {
    'C': [0.1, 1],
    'solver': ['liblinear'],
    'max_iter': [2000]
}

grid_search = GridSearchCV(
    estimator=LogisticRegression(
        class_weight='balanced',
        random_state=42
    ),
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    class_weight='balanced',
    solver='liblinear',
    max_iter=2000,
    random_state=42
)

model.fit(X_train, y_train)

print("✅ Model trained successfully!")
```

```python
y_pred = model.predict(X_test)

print("Predictions created!")
print(y_pred[:10])
```

```python
# Predict the class labels
y_pred = model.predict(X_test)

# Predict fraud probabilities (needed for ROC-AUC)
y_pred_proba = model.predict_proba(X_test)[:, 1]

print("Predictions generated successfully!")
print("First 10 Predictions:", y_pred[:10])
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("Accuracy :", round(accuracy,4))
print("Precision:", round(precision,4))
print("Recall   :", round(recall,4))
print("F1 Score :", round(f1,4))
print("ROC-AUC  :", round(roc_auc,4))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(7,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Non-Fraud','Fraud'],
    yticklabels=['Non-Fraud','Fraud']
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```

What the graph shows

The confusion matrix summarizes the classification performance by comparing actual transaction labels with predicted labels. It displays True Positives, True Negatives, False Positives, and False Negatives.

Why it matters

The confusion matrix provides a detailed understanding of model performance beyond overall accuracy. It helps identify whether the model is missing fraudulent transactions or generating too many false alarms.

Relation to fraud detection

In fraud detection, minimizing False Negatives is crucial because undetected fraudulent transactions can result in financial losses. At the same time, keeping False Positives low helps reduce unnecessary investigations of legitimate customers.

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8,6))

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()

plt.show()
```
