```python
X = df.drop("Class", axis=1)
y = df["Class"]

print("Features:", X.shape)
print("Target:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data:", X_test.shape)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
print(X_train.shape)
print(y_train.shape)
print(type(X_train))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {
    'C': [0.1, 1],
    'solver': ['liblinear'],
    'max_iter': [2000]
}

grid_search = GridSearchCV(
    estimator=LogisticRegression(
        class_weight='balanced',
        random_state=42
    ),
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1)

grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
print(hasattr(grid_search, "best_params_"))
```

```python
print("model exists:", "model" in globals())
print("grid_search exists:", "grid_search" in globals())
print("best_model exists:", "best_model" in globals())
print("y_pred exists:", "y_pred" in globals())
```

```python
hasattr(grid_search, "best_estimator_")
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    class_weight='balanced',
    solver='liblinear',
    max_iter=2000,
    random_state=42
)

model.fit(X_train, y_train)

print("✅ Model trained successfully!")
```

```python
y_pred = model.predict(X_test)

print("Predictions created!")
print(y_pred[:10])
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Non-Fraud', 'Fraud'],
    yticklabels=['Non-Fraud', 'Fraud']
)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

#Business Insights
The analysis confirms that credit card fraud is a highly imbalanced problem, with fraudulent transactions representing only a small fraction of all transactions. Logistic Regression provides a simple and interpretable baseline model for detecting suspicious activity. Evaluation metrics such as precision, recall, and the confusion matrix provide a better understanding of performance than accuracy alone. This type of model can assist financial institutions by flagging potentially fraudulent transactions for further investigation, helping reduce financial losses while supporting timely fraud prevention.
