```python
# Calculate Quartiles and IQR for the Amount column

Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Limit:", lower)
print("Upper Limit:", upper)
```

```python
outliers = df[(df['Amount'] < lower) | (df['Amount'] > upper)]

print("Number of outliers:", len(outliers))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    class_weight='balanced',
    solver='liblinear',
    max_iter=2000,
    random_state=42
)

model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
# Generate predictions
y_pred = model.predict(X_test)

print("Predictions created successfully!")
print(y_pred[:10])
```

```python
print("Model exists:", "model" in globals())
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Non-Fraud (0)', 'Fraud (1)'],
    yticklabels=['Non-Fraud (0)', 'Fraud (1)']
)

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.show()
```

#Business Insights
The analysis confirms that credit card fraud is a highly imbalanced problem, with fraudulent transactions representing only a small fraction of all transactions. Logistic Regression provides a simple and interpretable baseline model for detecting suspicious activity. Evaluation metrics such as precision, recall, and the confusion matrix provide a better understanding of performance than accuracy alone. This type of model can assist financial institutions by flagging potentially fraudulent transactions for further investigation, helping reduce financial losses while supporting timely fraud prevention.
